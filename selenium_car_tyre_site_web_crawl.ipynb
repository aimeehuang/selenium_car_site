```python
# webdriver連線:chrome要更新到最新版本
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from pyquery import PyQuery as pq 
from bs4 import BeautifulSoup
from bs4 import SoupStrainer
import time
import re
import pandas as pd
```

```python
chromedriver = "C:\webdriver\chromedriver"
driver = webdriver.Chrome(chromedriver)
driver.get(website link)


# 抓取下拉選單元件
select = Select(driver.find_element_by_id('car-make'))
time.sleep(1)


#找出所有品牌
brand =[]
Tyres = []
for op in select.options:
    brand.append(op.text)
#print(brand)
time.sleep(2)


for brands in brand[1:]:
    
    #點選品牌
    driver.find_element_by_id('car-make').click()
    select = Select(driver.find_element_by_id('car-make'))
    select.select_by_visible_text(str(brands))
    time.sleep(2)   


     #找出所有車款
    driver.find_element_by_id('car-model').click()
    time.sleep(2) 
    select = Select(driver.find_element_by_id('car-model'))   
    model =[]
    for op in select.options:
        model.append(op.text)
    #print(model)

    #select_element = WebDriverWait(driver, 30,0.5).until(EC.presence_of_element_located((By.ID, 'car-model')))
    #select = Select(select_element)

    #點選車款
    for models in model[1:]:
        print("model : "+models)
        driver.find_element_by_id('car-model').click()
        time.sleep(2) 
        select = Select(driver.find_element_by_id('car-model')) 

        select.select_by_visible_text(str(models))
        time.sleep(2) 

        #取得車型
        driver.find_element_by_id('car-trim').click()
        time.sleep(2) 
        select = Select(driver.find_element_by_id('car-trim'))  

        #印出所有車型
        trim =[]
        for op in select.options:
            trim.append(op.text)

        for trims in trim[1:]:
            print("trim : "+trims)
            time.sleep(3) 
            driver.find_element_by_id('car-trim').click()
            select = Select(driver.find_element_by_id('car-trim'))  
            select.select_by_visible_text(str(trims))
            time.sleep(3) 

            #點擊"重新搜尋"
            driver.find_element_by_xpath('//*[@id="search-by-trim"]/p[5]/input').click()
            time.sleep(3) 
            #only parse id=list elements
            soupStrainer = SoupStrainer(id="list")
            soup=  BeautifulSoup(driver.page_source,'html.parser', parse_only=soupStrainer)

            #找到目標元素
            tyre = soup.findAll('li',{'class':re.compile('^brand.*')})

            #剖析輪開款式及對應型號
            for tyres in tyre:
                name = tyres.findAll('h2',{'class':'title'})
                tyres =tyres.findAll('div',{'class':'detail'})
                #print(name[0].get_text())

                for eachtyre in tyres[0].findAll('li'):  
                    thisCar = []
                    thisCar.append(brands)
                    thisCar.append(models)
                    thisCar.append(trims)
                    thisCar.append(eachtyre.get_text())
                    Tyres.append(thisCar)

                print("----------Next----------")
```
